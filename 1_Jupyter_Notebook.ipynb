# Welcome to Jupyter Notebook!

A Jupyter Notebook is an open-source web application that allows you to create and share documents containing live code, mathematical equations, visualizations, and narrative text. It is a highly popular tool among data scientists, researchers, and developers for iterative coding, data analysis, and machine learning.

Instead of running a massive script all at once, a Jupyter Notebook breaks your work down into individual cells. This allows you to write, edit, and execute code in small, digestible chunks.

### Learning Objectives
1. Be familiar with the Jupyter Notebook environment.
2. Differentiate types of cells.
3. Be able to write using Markdown.

### Cells

There are three primary types of cells:
1. Markdown Cells: Allow you to write plain text formatted with Markdown. You use these to document your process, write explanations, and include bullet points or headings.
2. Code Cells: Contain actual executable code (e.g., Python). When you run a code cell, the output (data tables, graphs, or errors) appears directly below it.
3. Raw Cells: Used for writing raw text that does not get evaluated or formatted by the notebook. This is seldomly used.

To add a cell, be sure you're not typing on it or else press `Esc`, press `B` to add to the bottom or press `A` to add a cell above. Press `X` to delete a cell, be cautious though.

To change a cell type, you can select the drop down list under the Widgets tab. You can also press the following to change the cell type:
1. `M` for Markdown
2. `Y` for Code
3. `R` for Raw
